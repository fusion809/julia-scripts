# Lorenz attractor static plotting notebook
The [Lorenz equations](https://en.wikipedia.org/wiki/Lorenz_equations) are solved in this directory, namely:

$\begin{align*}
\dfrac{dx}{dt} &= \sigma (y-x) \\
\dfrac{dy}{dt} &= x(\rho-z) - y \\
\dfrac{dz}{dt} &= xy - \beta z
\end{align*}$

where $\sigma=10$, $\rho=28$ and $\beta=\dfrac{8}{3}$.

In this notebook, only static plots are shown; see [Lorenz animation.ipynb](Lorenz%20animation.ipynb) for animated plots. In this notebook, the initial conditions used are:

$\begin{align*}
x(0) &= -2.0 \\
y(0) &= 3.0  \\
z(0) &= 0.5.
\end{align*}$

The next cell uses the fourth order Runge-Kutta method to solve the problem.

```julia
include("RK4.jl")
```

The following script uses [ODE.jl](https://github.com/SciML/ODE.jl)'s `ode78` ODE solver to approximate the solution to the Lorenz equations. 

```julia
include("ode78.jl")
```
